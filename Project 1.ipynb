#### Target Budget

```python
budget_salary_income_list=[]
budget_bonus_income_list=[]
budget_other_income_list=[]
budget_housing_expense_list=[]
budget_education_expense_list=[]
budget_food_expense_list=[]
budget_auto_expense_list=[]
budget_health_expense_list=[]
budget_other_expense_list=[]

while True:
    print('Select Expected Budget section: ')
    print('1: Income')
    print('2: Expense')
    print('3: I am done creating my budget.')
    choice=int(input('Enter your choice (1-3): '))  
    if choice==1:
        while True:
            print('Select an income type:')
            print('1: Salary')
            print('2: Bonus')
            print('3: Other')
            budget_income=int(input('Enter income type here: '))
            
            if budget_income==1:
                budget_salary_income=int(input('Enter Salary amount expected: '))
                budget_salary_income_list.append(budget_salary_income)
                break
            elif budget_income==2:
                budget_bonus_income=int(input('Enter Bonus expected: '))
                budget_bonus_income_list.append(budget_bonus_income)
                break
            elif budget_income==3:
                budget_other_income=int(input('Enter Other income expected: '))
                budget_other_income_list.append(budget_other_income)
                break
            else:
                print('You chose a wrong option. Try again.')
    elif choice==2:
        while True:
            print('Select an expense type: ')
            print('1: Housing')
            print('2: Education')
            print('3: Food and Dining')
            print('4: Auto and Transportation')
            print('5: Health and Fitness')
            print('6: Other')
            
            budget_expense=int(input('Enter your expense type here: '))
            
            if budget_expense==1:
                budget_housing_expense=int(input('Enter Housing cost expected:'))
                budget_housing_expense_list.append(budget_housing_expense)
                break
            elif budget_expense==2:
                budget_education_expense=int(input('Enter Education cost expected:'))
                budget_education_expense_list.append(budget_education_expense)
                break
            elif budget_expense==3:
                budget_food_expense=int(input('Enter Food and Dining cost expected:'))
                budget_food_expense_list.append(budget_food_expense)
                break
            elif budget_expense==4:
                budget_auto_expense=int(input('Enter Auto and Transportation cost expected:'))
                budget_auto_expense_list.append(budget_auto_expense)
                break
            elif budget_expense==5:
                budget_health_expense=int(input('Enter Health and Fitness cost expected:'))
                budget_health_expense_list.append(budget_health_expense)
                break
            elif budget_expense==6:
                budget_other_expense=int(input('Enter Other cost expected:'))
                budget_other_expense_list.append(budget_other_expense)
                break
            else:
                print('You chose a wrong option. Try again.')            
    elif choice==3:
        print('You are done doing transactions! ')
        
        break
    else:
        print('You chose a wrong option. Try again.')


print(f'Total Salary Income expected: ${sum(budget_salary_income_list)}')
print(f'Total Bonus Income expected: ${sum(budget_bonus_income_list)}')
print(f'Total Other Income expected: ${sum(budget_other_income_list)}')

budget_total_income=[]
budget_total_income.append(sum(budget_salary_income_list))
budget_total_income.append(sum(budget_bonus_income_list))
budget_total_income.append(sum(budget_other_income_list))
print(f'Total Income expected: ${sum(budget_total_income)}')

print(f'Total Housing Expense expected: {sum(budget_housing_expense_list)}')
print(f'Total Education Expense expected: {sum(budget_education_expense_list)}')
print(f'Total Food and Dining Expense expected: {sum(budget_food_expense_list)}')
print(f'Total Auto and Transportation Expense expected: {sum(budget_auto_expense_list)}')
print(f'Total Health and Fitness Expense expected: {sum(budget_health_expense_list)}')
print(f'Total Other Expense expected: {sum(budget_other_expense_list)}')

budget_total_expense=[]
budget_total_expense.append(sum(budget_housing_expense_list))
budget_total_expense.append(sum(budget_education_expense_list))
budget_total_expense.append(sum(budget_food_expense_list))
budget_total_expense.append(sum(budget_auto_expense_list))
budget_total_expense.append(sum(budget_health_expense_list))
budget_total_expense.append(sum(budget_other_expense_list))
print(f'Total Expense expect: ${sum(budget_total_expense)}')

budget_net_value=[]
for i in range(len(budget_total_income)):
    budget_net_value.append(budget_total_income[i]-budget_total_expense[i])
print(f'Expected Budget Net Value: ${sum(budget_net_value)}')
```

#### Actual Budget

```python
salary_income_list=[]
bonus_income_list=[]
other_income_list=[]
housing_expense_list=[]
education_expense_list=[]
food_expense_list=[]
auto_expense_list=[]
health_expense_list=[]
other_expense_list=[]

while True:
    print('Select a transaction type: ')
    print('1: Income')
    print('2: Expense')
    print('3: No more transactions.')
    choice=int(input('Enter your choice (1-3): '))  
    if choice==1:
        while True:
            print('Select an income type:')
            print('1: Salary')
            print('2: Bonus')
            print('3: Other')
            income=int(input('Enter income type here: '))
            
            if income==1:
                salary_income=int(input('Enter Salary amount: '))
                salary_income_list.append(salary_income)
                break
            elif income==2:
                bonus_income=int(input('Enter Bonus here: '))
                bonus_income_list.append(bonus_income)
                break
            elif income==3:
                other_income=int(input('Enter Other income: '))
                other_income_list.append(other_income)
                break
            else:
                print('You chose a wrong option. Try again.')
    elif choice==2:
        while True:
            print('Select an expense type: ')
            print('1: Housing')
            print('2: Education')
            print('3: Food and Dining')
            print('4: Auto and Transportation')
            print('5: Health and Fitness')
            print('6: Other')
            expense=int(input('Enter your expense type here: '))
            
            if expense==1:
                housing_expense=int(input('Enter Housing cost: '))
                housing_expense_list.append(housing_expense)
                break
            elif expense==2:
                education_expense=int(input('Enter Education cost: '))
                education_expense_list.append(education_expense)
                break
            elif expense==3:
                food_expense=int(input('Enter Food and Dining cost: '))
                food_expense_list.append(food_expense)
                break
            elif expense==4:
                auto_expense=int(input('Enter Auto and Transportation cost: '))
                auto_expense_list.append(auto_expense)
                break
            elif expense==5:
                health_expense=int(input('Enter Health and Fitness cost: '))
                health_expense_list.append(health_expense)
                break
            elif expense==6:
                other_expense=int(input('Enter Other cost:'))
                other_expense_list.append(other_expense)
                break
            else:
                print('You chose a wrong option. Try again.')
    elif choice==3:
        print('You are done doing transactions!')
        
        break
    else:
        print('You chose a wrong option. Try again.')

print(f'Total Salary Income: ${sum(salary_income_list)}')
print(f'Total Bonus Income: ${sum(bonus_income_list)}')
print(f'Total Other Income: ${sum(other_income_list)}')

total_income=[]
total_income.append(sum(salary_income_list))
total_income.append(sum(bonus_income_list))
total_income.append(sum(other_income_list))
print(f'Total Income: ${sum(total_income)}')

print(f'Total Housing Expense: {sum(housing_expense_list)}')
print(f'Total Education Expense: {sum(education_expense_list)}')
print(f'Total Food and Dining Expense: {sum(food_expense_list)}')
print(f'Total Auto and Transportation Expense: {sum(auto_expense_list)}')
print(f'Total Health and Fitness Expense: {sum(health_expense_list)}')
print(f'Total Other Expense: {sum(other_expense_list)}')

total_expense=[]
total_expense.append(sum(housing_expense_list))
total_expense.append(sum(education_expense_list))
total_expense.append(sum(food_expense_list))
total_expense.append(sum(auto_expense_list))
total_expense.append(sum(health_expense_list))
total_expense.append(sum(other_expense_list))
print(f'Total Expense: ${sum(total_expense)}')

net_value=[]
for i in range(len(total_income)):
    net_value.append(total_income[i]-total_expense[i])
print(f'Net Value: ${sum(net_value)}')
```

#### Budget Variance

```python
print(f'Total Income expected: ${sum(budget_total_income)}')
print(f'Total Income: ${sum(total_income)}')
print(f'Income Variance: ${sum(total_income)-sum(budget_total_income)}.')

while True:
    if sum(total_income)>sum(budget_total_income):
        print(f'Congratulations! Your income was higher than expected by ${sum(total_income)-sum(budget_total_income)}!')
        break
    elif sum(total_income)<sum(budget_total_income):
        print(f' Too bad! You made less than expected by ${sum(total_income)-sum(budget_total_income)}.')
        break
    else:
        print('You made as much expected!')
        break

print(f'Total Expense expect: ${sum(budget_total_expense)}')
print(f'Total Expense: ${sum(total_expense)}')
print(f'Expense Variance: ${sum(total_expense)-sum(budget_total_expense)}.')

while True:
    if sum(total_expense)>sum(budget_total_expense):
        print(f'Too bad! You spent more than expected by ${sum(total_expense)-sum(budget_total_expense)}.')
        break
    elif sum(total_expense)<sum(budget_total_expense):
        print(f' Congratulations! Your expenditure was less than expected by ${sum(total_expense)-sum(budget_total_expense)}!')
        break
    else:
        print('You spent as much expected!')
        break


print(f'Expected Budget Net Value: ${sum(budget_net_value)}')
print(f'Net Value: ${sum(net_value)}')
print(f'Net Value Variance: ${sum(budget_net_value)-sum(net_value)}.')

while True:
    if sum(net_value)>0:
        print(f'Congratulations! You have ${sum(net_value)} to spare. What do you want to do with it?')
        print('1: Invest 25% and save 75%')
        print('2: Invest 50% and save 50%')
        print('3: Invest 75% and save 25%')
        print('4: Nothing. I got it from here.')
        choice=int(input('Enter your choice (1-3): '))
        if choice==1:
            invest_1=0.25*sum(net_value)
            saving_1=0.75*sum(net_value)
            print(f'You are now investing ${invest_1} and saving ${saving_1}. Thanks!')
            break
        elif choice==2:
            invest_2=0.50*sum(net_value)
            saving_2=0.50*sum(net_value)
            print(f'You are now investing ${invest_2} and saving ${saving_2}. Thanks!')
            break
        elif choice==3:
            invest_3=0.75*sum(net_value)
            saving_3=0.25*sum(net_value)
            print(f'You are now investing ${invest_3} and saving ${saving_3}. Thanks!')
            break
        elif choice==4:
            print('Thank you and good luck!')
            break
        else:
            print('You chose a wrong option. Try again.')
            break
                  
    elif sum(net_value)<0:
        print(f'Oh no! You are in the negatives by ${sum(net_value)}. Reduce expenses!')
        break
```
